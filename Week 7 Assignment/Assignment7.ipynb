# Read the Data

```python
import pandas as pd
try:
    df = pd.read_csv("iris.csv")
except FileNotFoundError:
    print("Error: The file was not found")
except pd.errors.EmptyDataError:
    print("Error: The file is empty.")
except pd.errors.ParserError:
    print("Error: The file is not a valid CSV or is malformed.")
except Exception as e:
    print(f"Unexpected error: {e}")

df.head()    #View the first five rows of the dataset
```

# Pie chart of categorical variable

```python
import matplotlib.pyplot as plt
species_counts = df['Species'].value_counts()  #get count of each species
plt.figure(figsize=(6, 6))
plt.pie(
    species_counts,
    labels=species_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Pastel1.colors
)
plt.title("Distribution of Iris Species")
plt.axis('equal') 
```

# Plot Histogram of Sepal Length

```python
import matplotlib.pyplot as plt

# Plot histogram
plt.figure(figsize=(6, 4))
plt.hist(df['SepalLengthCm'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# Scatterplot for Petal Length against Sepal Length

```python
import matplotlib.pyplot as plt
import numpy as np

x = df['SepalLengthCm']
y = df['PetalLengthCm']

slope, intercept = np.polyfit(x, y, deg=1)
regression_line = slope * x + intercept

plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='teal', edgecolor='black')
plt.plot(x, regression_line, color='red', label='Regression Line')

equation_text = f'y = {slope:.2f}x + {intercept:.2f}'
plt.text(x.min() + 0.2, y.max() - 0.1, equation_text, fontsize=12, color='red')  # Add regression equation

plt.title('Petal Length vs Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
